```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
def split(x):
    if ' ' in x:
        y = x.split(' ')[0]
    elif 'T' in x:
        y = x.split('T')[0]
    return y
```

```python
def fix(x):
    x = split(x)
    if '/' in x:
        if '2020' in x:
            y = datetime.strptime(x, '%m/%d/%Y').date()
        else:
            y = datetime.strptime(x, '%m/%d/%y').date()
    elif '-' in x:
        if '2020' in x:
            y = datetime.strptime(x, '%Y-%m-%d')
        else:
            y = datetime.strptime(x, '%y-%m-%d').date()
    return y
```

```python
pth = '../US_covid_cases.csv'
us = pd.read_csv(pth)
```

```python
'WA' in 'Seattle, WA'
```

```python
wash = us[(us['Province_State'] == 'Washington') | (us['Province_State'].str.contains('WA'))]
```

```python
wash
```

```python
wash = wash[['Province_State', 'Confirmed', 'Deaths', 'Recovered', 'Last_Update']]
```

```python
wash['Date'] = wash['Last_Update'].apply(fix)
```

```python
wash
```

```python
wash = wash.drop_duplicates(['Province_State', 'Last_Update', 'Date'])
```

```python
wash.count()
```

```python
wash = wash.fillna(0)
wash = wash.groupby('Date').sum()
```
